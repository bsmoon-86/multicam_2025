# 🧮 Python 클래스 연습 문제 – 직사각형 & 학생 클래스
이 노트북은 **Python 클래스(class)** 기초 개념을 연습하기 위한 예제입니다.

아래에 두 가지 클래스를 구현합니다.
1. `Rectangle` 클래스 – 직사각형의 정보를 관리하고 넓이/둘레를 계산
2. `Student` 클래스 – 학생의 점수를 관리하고 평균/등급을 계산


## ✅ 문제 1: 직사각형(Rectangle) 클래스

다음 요구사항에 맞는 `Rectangle` 클래스를 작성하세요.

### 📌 요구사항
1. **객체 생성 시 가로(width), 세로(height)를 입력** 받아 속성으로 저장한다.
   - 예: `rect = Rectangle(10, 5)` → 가로 10, 세로 5인 직사각형
2. `rectangle_info()` 메서드를 구현한다.
   - 반환 형식: `"가로: X, 세로: Y"`
   - 예: 가로 10, 세로 5라면 → `"가로: 10, 세로: 5"`
3. `area()` 메서드는 **직사각형의 넓이**를 반환한다.
   - 공식: `가로 * 세로`
   - 예: 가로 10, 세로 5 → `50`
4. `perimeter()` 메서드는 **직사각형의 둘레**를 반환한다.
   - 공식: `2 * (가로 + 세로)`
   - 예: 가로 10, 세로 5 → `30`


```python
# 문제 1 정답 코드: Rectangle 클래스 구현

class Rectangle:
    def __init__(self, width, height):
        """가로(width)와 세로(height)를 받아 직사각형을 초기화한다.
        width 또는 height가 0 이하인 경우 예외를 발생시킨다.
        """
        if width <= 0 or height <= 0:
            raise ValueError("width와 height는 0보다 커야 합니다.")
        self.width = width
        self.height = height

    def rectangle_info(self):
        """직사각형의 가로와 세로 정보를 문자열로 반환한다."""
        return f"가로: {self.width}, 세로: {self.height}"

    def area(self):
        """직사각형의 넓이를 반환한다."""
        return self.width * self.height

    def perimeter(self):
        """직사각형의 둘레를 반환한다."""
        return 2 * (self.width + self.height)


# 사용 예시
rect = Rectangle(10, 5)
print(rect.rectangle_info())   # 가로: 10, 세로: 5
print("넓이:", rect.area())    # 넓이: 50
print("둘레:", rect.perimeter())  # 둘레: 30
```

## ✅ 문제 2: 학생(Student) 클래스

다음 요구사항에 맞는 `Student` 클래스를 작성하세요.

### 📌 요구사항
1. **객체 생성 시 이름(name)을 입력** 받아 속성으로 저장한다.
   - 예: `st = Student("홍길동")`
2. `scores` 라는 **리스트 속성**을 가지고, 학생의 점수들을 저장한다.
   - 초기에는 빈 리스트로 시작한다. (`[]`)
3. `add_score(score)` 메서드를 구현한다.
   - 매개변수 `score`를 `scores` 리스트에 추가한다.
4. `average()` 메서드를 구현한다.
   - 현재까지 저장된 모든 점수의 **평균을 소수점 둘째 자리까지 반올림**하여 반환한다.
   - 만약 `scores`가 비어 있다면, 문자열 `"점수가 없습니다."` 를 반환한다.
5. `grade()` 메서드를 구현한다.
   - 평균 점수에 따라 아래 기준으로 등급(A~F)을 반환한다.

   | 평균 점수 | 등급 |
   |----------|------|
   | 90 ~ 100 | A |
   | 80 ~ 89  | B |
   | 70 ~ 79  | C |
   | 60 ~ 69  | D |
   | 0 ~ 59   | F |

   - 만약 점수가 하나도 없는 경우(= `scores`가 비어 있는 경우)에는
     - 문자열 `"등급 없음"` 을 반환한다.

6. `info()` 메서드를 구현한다.
   - 아래 형식의 문자열을 반환한다.
   - `"학생 이름: OOO, 점수: [점수 리스트], 평균: X점, 등급: Y"`
   - 예: 점수가 [95, 88, 72]이고 평균이 85.0, 등급이 B라면
     - `"학생 이름: 홍길동, 점수: [95, 88, 72], 평균: 85.0점, 등급: B"`



```python
# 문제 2 정답 코드: Student 클래스 구현

class Student:
    def __init__(self, name):
        """학생 이름을 받아 초기화하고, 점수 리스트를 빈 리스트로 생성한다."""
        self.name = name
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)

    def average(self):
        """점수들의 평균을 소수 둘째 자리까지 반올림하여 반환한다.
        점수가 없다면 '점수가 없습니다.'를 반환한다.
        """
        if not self.scores:
            return "점수가 없습니다."
        avg = sum(self.scores) / len(self.scores)
        return round(avg, 2)

    def grade(self):
        """평균 점수에 따른 등급(A~F)을 반환한다.
        점수가 없으면 '등급 없음'을 반환한다.
        """
        if not self.scores:
            return "등급 없음"
        avg = self.average()
        if avg >= 90:
            return "A"
        elif avg >= 80:
            return "B"
        elif avg >= 70:
            return "C"
        elif avg >= 60:
            return "D"
        else:
            return "F"

    def info(self):
        """학생의 이름, 점수 목록, 평균, 등급 정보를 문자열로 반환한다."""
        return (
            f"학생 이름: {self.name}, 점수: {self.scores}, "
            f"평균: {self.average()}점, 등급: {self.grade()}"
        )


# 사용 예시
st = Student("홍길동")
st.add_score(95)
st.add_score(88)
st.add_score(72)
print(st.info())
```
