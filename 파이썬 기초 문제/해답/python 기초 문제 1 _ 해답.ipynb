# 문제3
- 2부터 50까지의 숫자들중 소수를 리스트의 형태로 출력하시오
    - 소수는 1과 자기 자신 외에는 나누어지지 않는 수이다. 
    - 2부터 50까지 반복을 하면서 2부터 시작하여 자기 자신의 -1 까지 나누어 나머지 값이 0 존재하지 않는 수들을 리스트에 담아서 출력하시오

```python
cnt = 0
for num in range(2, 51):  # 2부터 50까지 반복
    is_prime = True       # 소수라고 가정
    
    for i in range(2, num):
        if num % i == 0:  # 나누어 떨어지면 소수가 아님
            is_prime = False
            cnt += 1
            break

    if is_prime:
        print(num, end=" ")
print(cnt)
```

- 음수형 데이터도 적용이 가능하도록 코드를 작성하시오
- 예시)

    - 입력값 : -9876, 출력값 : -6789

```python
n = 12345
# range(시작값, 종료값, 증감값)
# [ 시작 위치 : 종료 위치 : 단계(기본값1)]
# str(n)[::-1]
# 매개변수 -> 함수가 호출될때마다 변할 수 있는 데이터 값
def reversed(n):
    if n >= 0:
        str_n = str(n)
        result = ""
    else:
        #  숫자에서 절대값 (abs())
        str_n = str(abs(n))
        result = '-'
    # for i in range(len(str_n)-1, -1, -1):
    #     # str_n의 idx -> 0 ~ 4
    #     print(str_n[i])

    for i in range(1, len(str_n)+1):
        # str_n의 idx -> 0 ~ 4
        # idx에서 음수의 의미는 뒤에서부터
        # print(str_n[-i])
        result += str_n[-i]
    print(result)
```

```python
# 함수 -> def 키워드를 이용하여 함수의 정의 
# def 함수명(매개변수):             매개변수의 개수는 가변(0~N)
    # 함수에서 실행되는 코드 
    # return 반환값

# 함수 호출 -> 함수명(인자)          매개변수의 개수만큼 인자의 개수도 입력
# 함수는 행동(데이터를 변환)
# return : 되돌려주다. (값을 반환 -> 함수를 호출한 뒤 결과를 받아오는 경우)
# 함수를 하나만 사용하는 경우는 드물게 일어난다. 
# 하나의 함수를 호출하고 결과 값을 이용하여 다음 함수에 대입 -> 최종의 결과물을 확인
```

```python
def func_1(input_data):
    # input_data를 절대값으로 변환 result에 대입
    result = abs(input_data)
    # result를 역순으로 변경 -> 숫자형 데이터를 문자로 변환 역순으로 변환 다시 숫자로 변경
    result = int(str(result)[::-1])
    if input_data < 0:
        # input_data가 음수라면 result에 -1을 곱하여 반환
        result = result*-1
    return result

print(func_1(12345))
print(func_1(-9876))
print(func_1(-112233))
print(func_1(0))

# 함수는? -> 
# 1. func_1  : 함수 이름    -> 라이브러리 호출 -> from 라이브러리명 import 함수명
# 2. func_1()  : 함수 호출
# 3. def func_1() : 함수 정의
```

# 문제7
- 특정 숫자를 입력하면 2진수로 변환을 하는 함수를 생성하시오
- 매개변수는 1개로 함수 호출 시 인자 값을 2진수로 변환하여 되돌려주는 함수를 생성하시오
- 인자 값이 0인 경우에는 0을 되돌려주고 양수인 경우와 음수인 경우를 모두 생각해서 함수를 구성하시오

```python
# 무한 반복문을 이용하여 입력 받은 데이터를 2로 나눠준다. 
# 몫이 0이 되는 순간 반복문을 종료(break)
n = -10         # 함수에서 입력받은 데이터 
def func_2(n):
    result = ''
    n2 = abs(n)
    while True:
        # n을 2로 나눈 나머지(type? int -> str)를 어딘가(result)에 저장 
        result += str(n2 % 2)
        # result = result + str(n % 2)
        # str + int 에러 발생
        # n을 2로 나눈 몫
        n2 = n2 // 2
        # print(n)
        if n2 == 0:
            # 2로 나눈 몫이 0이 되면 반복문 종료
            break
    # result를 역순 -> 2진수 변환 완료 -> 음수라면 앞에 - 더해준다. 
    if n >= 0:
        return result[::-1]         # 함수에서는 return
    else:
        return '-' + result[::-1]   # 함수에서는 return

print(func_2(10))
print(func_2(-1000))
```

```python
def reverse_list(_list):
    list_len = len(_list)
    for i in range(list_len//2):
        _list[i], _list[- i - 1] = _list[- i - 1], _list[i]
    return _list


reverse_list([1,2,3,4,5])
reverse_list([1,2,3,4,5, 6, 7, 8, 9])
# reverse_list('apple')
```

# 문제9
- 특정 숫자를 연속된 자연수의 합을 출력하는 함수를 생성하시오
- 인자값을 받아와서 연속된 자연수의 합이 인자값과 같은 모든 경우의 수를 리스트 형태로 출력

```python
for i in range(1, 15):
    # 누적합의 초기값(sum_init) 생성
    # 빈 리스트(_list) 생성
    for j in range():
        # j의 값을 sum_init에 누적 합
        # j의 값을 _list에 추가
        # sum_list가 15가 된다면 리스트를 출력을 하고 반복문을 종료 
        # sum_list가 15보다 커지면 반복문을 종료


        # 누적합    ->  누적 합을 구할수 있는 공간을 생성 
        # 반복되는 리스트   -> 반복되는 숫자들을 추가할 수 있는 공간을 생성
        # 누적합이 15가 된다면 리스트를 출력을 하고 반복문을 종료 
        # 누적합이 15보다 커지면 반복문을 종료
```

```python

def consecutive_sums(n):
    results = []

    # 연속된 숫자의 시작값 i를 1부터 n까지 탐색
    for i in range(1, n):
        # 1부터 14까지 i에 값을 대입하여 반복
        total = 0
        seq = []
        # 값들을 초기화
        # i가 1이 대입 
        # total = 0,seq = []
        # i부터 계속 더하며 n과 같은지 비교
        for j in range(i, n):
            # i부터 14까지 j에 값을 대입하여 반복
            # j에 1이 대입 
            # total += 1 -> total(1)
            # seq -> [1]
            # total이 15와 같은가? , total이 15보다 큰가? -> 모두 거짓 -> 반복문은 유지
            # j에 2가 대입 
            # total(1) += 2 -> total(3)
            # seq -> [1,2]
            # total이 15와 같은가?, total이 15보다 큰가? -> 모두 거짓 -> 반복문은 유지
            # j에 3이 대입 
            # total(3) += 3 -> total(6)
            # seq -> [1,2,3]
            # total이 15와 같은가?, total이 15보다 큰가? -> 모두 거짓 -> 반복문은 유지
            # j에 4이 대입 
            # total(6) += 4 -> total(10)
            # seq -> [1,2,3,4]
            # total이 15와 같은가?, total이 15보다 큰가? -> 모두 거짓 -> 반복문은 유지
            # j에 5이 대입 
            # total(10) += 5 -> total(15)
            # seq -> [1,2,3,4,5]
            # total이 15와 같은가?, total이 15보다 큰가? -> 모두 참 -> 2번째 반복문은 종료
            # total에 j를 누적합
            total += j
            # j의 값을 seq에 추가
            seq.append(j)

            if total == n:
                results.append(seq.copy())
                break
            elif total > n:
                break

    return results


print(consecutive_sums(15))
print(consecutive_sums(55))
```

# 문제10
- 두 개의 단어가 애너그램인지 판단을 하는 함수를 생성하시오
- 두개의 매개변수에 입력이 되는 두 인자의 값이 애너그램인지 bool 형태로 출력하시오

```python
def is_anagram(s1, s2):
    # 소문자로 변환 + 공백 제거 (문장일 수도 있으니)
    s1 = s1.replace(" ", "").lower()
    s2 = s2.replace(" ", "").lower()

    # 정렬한 결과가 같으면 애너그램
    return sorted(s1) == sorted(s2)

print(is_anagram("listen", "silent")) 
print(is_anagram("hello", "world"))  
```
