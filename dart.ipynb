```python
def load_dart_data( corp_code, year, reprt_code ):
    # corp_code : 종목코드
    # year : 사업연도
    # reprt_code : 보고서 구분
    url = 'https://opendart.fss.or.kr/api/fnlttSinglAcnt.json'
    params = {
        'crtfc_key' : key, 
        'corp_code' : corp_code,   ## 종목 코드 선택
        'bsns_year' : year,       ## 사업연도
        'reprt_code' : reprt_code     ## 보고서 구분 : 1분기(11013), 반기(11012), 3분기(11014), 사업보고서(11011)
    }
    response = requests.get(url, params=params)
    data = response.json()['list']
    result = pd.DataFrame(data)
    return result
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

# 1. 변수가 1개인 데이터 (예: 값이 '3' 하나뿐인 데이터)
X_single = np.array([[3]]) 

# degree=3 (1차, 2차, 3차식까지 생성)
poly3 = PolynomialFeatures(degree=3, include_bias=False)
result_single = poly3.fit_transform(X_single)

print("--- 변수 1개, degree=3 변환 ---")
print("원본 데이터:", X_single[0])
print("변환 결과 (x, x^2, x^3):", result_single[0]) 
# 출력 결과: [3.  9. 27.] -> 3^1, 3^2, 3^3 이 한 번에 만들어짐!


# 2. 변수가 2개인 데이터 (예: [공부 2시간, 수면 4시간])
X_multi = np.array([[2, 4]])

# degree=2 (각각의 1/2차식, 그리고 둘의 곱셈 교차항 생성)
poly2 = PolynomialFeatures(degree=2, include_bias=False)
result_multi = poly2.fit_transform(X_multi)

print("\n--- 변수 2개, degree=2 변환 ---")
print("원본 데이터 (x1, x2):", X_multi[0])
print("변환 결과 (x1, x2, x1^2, x1*x2, x2^2):", result_multi[0])
# 출력 결과: [ 2.  4.  4.  8. 16.] -> (2, 4) 그리고 (2^2, 2*4, 4^2) 생성!
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 1. 2개의 변수를 가진 데이터 준비 (X: [공부시간, 모의고사횟수])
# y: 최종 수능 점수 (두 변수가 결합할 때 점수가 기하급수적으로 오른다고 가정)
X = np.array([[2, 1], [3, 2], [4, 3], [5, 4]])
y = np.array([50, 65, 85, 100])

# 2. 다중 다항 변환 (degree=2)
# x1, x2를 넣으면 -> 1, x1, x2, x1^2, x1*x2, x2^2 로 변환해줍니다.
poly = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly.fit_transform(X)

print("--- 데이터 변환 과정 ---")
print(f"원본 데이터 (x1, x2): {X[0]}")
print(f"변환된 데이터 (x1, x2, x1^2, x1*x2, x2^2): {X_poly[0]}")

# 3. 모델 학습
model = LinearRegression()
model.fit(X_poly, y)

print("\n--- 학습 결과 ---")
print("찾아낸 가중치들:", np.round(model.coef_, 2))
```
