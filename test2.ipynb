```python
import pandas as pd 
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from sqlalchemy import create_engine
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://comp.wisereport.co.kr/company/c1010001.aspx')
```

```python
html_data = driver.page_source
```

```python
soup = bs(html_data, 'html.parser')
```

```python
div_list = soup.find_all(
    'div', 
    attrs = {
        'class' : 'fl_le'
    }
)
```

```python
len(div_list)
```

```python
list(
    map(
        lambda x : '주요지표' in x.get_text() or '어닝서프라이즈' in x.get_text(), 
        div_list
    )
)
```

```python
pd.read_html(str(div_list[1]) )[0]
```

```python
df1 = pd.read_html(str(div_list[1]))[0]
df1
```

```python
df2 = pd.read_html(str(div_list[2]))[0]
df2
```
