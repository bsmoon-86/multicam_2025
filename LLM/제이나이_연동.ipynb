```python
# 가장 기본적인 형태 

import os
from google import genai

# 제미나이 api 키 등록
client = genai.Client(api_key='')

# 모델 선택
resp = client.models.generate_content(
    # 모델의 버전을 선택
    model="gemini-2.5-flash",
    # 모델에 보내는 메시지 지정
    contents="카드 사용 데이터로 고객 이탈을 예측할 때 중요한 피처 5가지를 쉬운 말로 설명해줘."
)
# resp는 응답 메시지를 출력
print(resp.text)
```

```python
# 실시간으로 생성되는 텍스트들을 바로바로 출력하는 방법 (stream 방식 -> SSE 방식)
# 일반적으로 가장 많이 사용이 되는 형태 
client = genai.Client(api_key='')

stream = client.models.generate_content_stream(
    model="gemini-2.5-flash",
    contents="고객 이탈 설명 리포트 템플릿을 항목별로 만들어줘."
)

for chunk in stream:
    if chunk.text:
        print(chunk.text, end="")
```
