```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/9_1_2.csv')
df.head()
```

```python
df1 = df[df["구분"] == "검거건수"].set_index("연도").drop(columns="구분")
df2 = df[df["구분"] == "발생건수"].set_index("연도").drop(columns="구분")
```

```python
df1.head()
```

```python
dfratio = df1 / df2
```

```python
dfratio
```

```python
max_data = []
for idx in dfratio.index:
    max_value = max(dfratio.loc[idx])
    for col, val in zip(dfratio.columns , dfratio.loc[idx]):
        # print(col)
        # print(val)
        if val == max_value:
            max_data.append( [idx, col] )
max_data
```

```python
max_df = pd.DataFrame(max_data, columns = ['연도', '범죄유형'])
max_df
```

```python
final_data = []
for year, kind in max_df.values:
    # print(year)
    # print(kind)
    val = df1.loc[year, kind]
    final_data.append(
        [year, kind, val]
    )
final_df = pd.DataFrame(final_data, columns = ['연도', '범죄유형', '검거건수'])
```

```python
final_df.head()
```

```python
final_df['검거건수'].sum()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/9_1_3.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df["평균만족도"] = df["평균만족도"].fillna(df["평균만족도"].mean())
```

```python
df.isna().sum()
```

```python
mean_tenure = (
  df.groupby(["부서", "등급"])["근속연수"]
  .mean()
  .astype(int)
  .reset_index()
  .rename(columns = {"근속연수": "평균근속연수"})
)
```

```python
mean_tenure
```

```python
df = pd.merge(df, mean_tenure, on=['부서', "등급"], how='left')
df.head()
```

```python
flag_null = df['근속연수'].isna()
```

```python
df.loc[flag_null, '근속연수'] = df.loc[flag_null, '평균근속연수']
```

```python
df.loc[flag_null]
```

```python
df.isna().sum()
```

```python
A = df.loc[
    (df['부서'] == 'HR') & (df['등급'] == 'A'), 
    '근속연수'
].mean()
```

```python
A
```

```python
B = df.loc[
    (df['부서'] == 'Sales') & (df['등급'] == 'B'), 
    '교육참가횟수'
].mean()
```

```python
B
```
