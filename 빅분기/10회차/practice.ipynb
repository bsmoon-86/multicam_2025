# 1번 문제 
- subject-performance.csv 파일을 로드 '
- 소주제별 정답율을 계산
    - 정답률 = (해당 소주제의 정답 개수) / (해당 소주제의 전체 문제 풀이 수)
- 소주제들의 정답률 중 3번째로 높은 값을 구하시오
    - 정답률이 같다면 하나로 취급한다.
- 제출 형식은 소수점 셋째 자리까지 작성

```python
import pandas as pd
```

```python
df = pd.read_csv("data/subject_performance.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# 정답 컬럼의 데이터가 어떤 데이터가 존재하는가 확인 
df['is_correct'].unique()
```

```python
# 해당 데이터는 정답(1)과 오답(0)으로 나눠져있음
# 정답률은 소주제별 그룹화를 하고 평균을 구하면 된다. 
group_data =  df.groupby(['sub_topic'])['is_correct'].mean()
```

```python
group_data
```

```python
group_data = group_data.sort_values(ascending=False)
```

```python
group_data
```

```python
# 중복 값을 하나로 취급하니 중복 데이터를 제거 
group_data.drop_duplicates(inplace=True)
```

```python
# 3번째 값을 소수점 3번째 자리를 유지하고 반올림
round(group_data.iloc[2], 3)
```

```python
# 그룹화 -> 정답의 개수 / 총 데이터의 개수
# 분자에 들어갈 소주제별 정답의 개수
group_data2 = df.groupby(['sub_topic'])['is_correct'].sum()
# 분모에 들어갈 소주제별 문제의 개수 case1
group_data3 = df.groupby(['sub_topic'])['is_correct'].count()
```

```python
# 분모에 들어갈 소주제별 문제의 개수 case2
group_data4 = df['sub_topic'].value_counts()
```

```python
group_data
```

```python
group_data2 / group_data4
```

```python
years, months = [], []
# 연월 컬럼을 생성 -> 문자에서 앞의 n개의 글자를 추출
for date in df2['order_date']:
    split_str = date.split('.')
    years.append(split_str[0])
    months.append(split_str[1])
    # print(f"Year: {split_str[0]}, Month: {split_str[1]}")
test_df = df2.copy()
test_df['year'] = years
test_df['month'] = months
test_df.head()
```

- 스팸 메시지와 정상 메시지의 평균 단어의 개수를 계산
- 두 평균의 차이를 절대값으로 구하고 출력
- 제출형식 : 소수점 3번째 자리까지 유지하고 반올림

```python
df3 = pd.read_csv("data/hamspam.csv")
df3.info()
```

```python
df3.head()
```

```python
# for text in df3['text']:
#     print(len(text.split()))

[ len(text.split()) for text in df3['text'] ]
```

```python
df3['cnt'] = df3['text'].str.split().map(len)
```

```python
df3
```

```python
group_data3 =  df3.groupby('label')['cnt'].mean()
```

```python
group_data3
```

```python
abs(
    round(
        group_data3[0] - group_data3[1], 
        3
    )
)
```

```python
ham_df = df3.loc[
    df3['label'] == 'ham', 
]
spam_df = df3.loc[
    df3['label'] == 'spam', 
]

```

```python
ham_mean = ham_df['cnt'].mean()
spam_mean = spam_df['cnt'].mean()

print(f"ham mean: {ham_mean}")
print(f"spam mean: {spam_mean}")
```

```python
abs(
    round(
        ham_mean - spam_mean, 3
    )
)
```
